# Morocco in The Olympics

```python
import numpy as np
import pandas as pd
```

```python
results = pd.read_csv('results (2).csv')
```

```python
results.head()
```

### How many medals has Morocco won?

```python
medals = results[(results.medal.notna()) &  (~results['event'].str.endswith('(YOG)'))]
medals.drop_duplicates(subset=['year', 'type', 'discipline', 'event', 'noc', 'medal'], inplace=True)
medalPerCountry = medals.groupby('noc').medal.value_counts().reset_index()
medalPerCountry[medalPerCountry.noc == 'MAR']
```

### In which disciplines has Morocco won its medals?

```python
medals[medals.noc=='MAR'].groupby('discipline').medal.value_counts()
```

### Morocco's olympic participations per discipline?

```python
participations = results[~results['event'].str.endswith('(YOG)')]
participations.drop_duplicates(subset=['year', 'type', 'discipline', 'event', 'noc', 'medal'], inplace=True)
participations_MAR = participations[participations.noc=='MAR']
```

```python
participations_MAR.discipline.value_counts().reset_index()
```

### Morocco's participations over the years

```python
PartPerYear = participations_MAR.groupby(['year', 'type']).year.value_counts().reset_index()
PartPerYear
```

### Winter vs summer games participations

```python
#winterG = participations[(participations.noc=='MAR') & (participations.type == "Winter")]
winterG = participations_MAR[participations_MAR.type == "Winter"]
winterG.groupby('year').discipline.value_counts().reset_index()
```

```python
winterG.groupby('year').medal.count().reset_index()
```

```python
SummerG = participations[(participations.noc=='MAR') & (participations.type == "Summer")]
SummerG.groupby('year').discipline.value_counts().reset_index()
```

```python
SummerG.groupby('year').medal.count().reset_index()
```
